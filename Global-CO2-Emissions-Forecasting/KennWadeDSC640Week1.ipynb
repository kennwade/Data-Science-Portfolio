## Dataset Description

For this assignment, I have used the Titanic dataset, which contains information about the passengers who traveled on the Titanic. The dataset includes passenger class, name, sex, age, number of siblings/spouses aboard, number of parents/children aboard, and fare paid.

The dataset consists of the following fields:
1. `Survived`: Indicates if the passenger survived (1) or not (0).
2. `Pclass`: Passenger class (1 = 1st, 2 = 2nd, 3 = 3rd).
3. `Name`: Name of the passenger.
4. `Sex`: Sex of the passenger.
5. `Age`: Age of the passenger.
6. `Siblings/Spouses Aboard`: Number of siblings/spouses aboard.
7. `Parents/Children Aboard`: Number of parents/children aboard.
8. `Fare`: Fare paid by the passenger.

The dataset can be found on Kaggle at the following URL:
[https://www.kaggle.com/c/titanic/data](https://www.kaggle.com/c/titanic/data)

## Assignment Instructions

For this assignment, I will create the following visualizations using Python:
1. **Bar Chart**: Count of passengers by class.
2. **Stacked Bar Chart**: Count of passengers by class and sex.
3. **Pie Chart**: Distribution of passengers by class.
4. **Donut Chart**: Distribution of passengers by class.
5. **Line Chart**: Relationship between age and fare by class.

These visualizations will help me understand the distribution and demographics of the passengers on the Titanic.


```python
# Step 1: Import Libraries
print("Step 1: Importing necessary libraries...")
import matplotlib.pyplot as plt  # Library for creating visualizations
import seaborn as sns  # Library for creating statistical graphics
import pandas as pd  # Library for data manipulation and analysis
print("Libraries imported successfully!")
```

```python
# Step 2: Load Dataset from Local File
print("Step 2: Loading dataset from local file...")
# Load the dataset from the specified CSV file path
file_path = '/Users/kennwade/Desktop/640 Data Presentation /titanic.csv'
data = pd.read_csv(file_path)
print("Dataset loaded successfully!")
print(data.head())  # Display the first few rows of the dataset to confirm
```

```python
# Step 3: Create a Bar Chart
print("Step 3: Creating a bar chart...")
plt.figure(figsize=(10, 6))  # Set the size of the figure
sns.countplot(data=data, x='Pclass')  # Create a count plot (bar chart) for the 'Pclass' column
plt.title('Python - Bar Chart - Count of Passengers by Class')  # Set the title of the chart
plt.xlabel('Class')  # Set the label for the x-axis
plt.ylabel('Count of Passengers')  # Set the label for the y-axis
plt.savefig('Python_Bar_Chart.png')  # Save the figure as a PNG file
plt.show()  # Display the chart
print("Bar chart created and saved as 'Python_Bar_Chart.png'!")
```

```python
# Step 4: Create a Stacked Bar Chart
print("Step 4: Creating a stacked bar chart...")
# Pivot the dataset to get the count of survived passengers by class and sex
pivot_table = data.pivot_table(index='Pclass', columns='Sex', values='Survived', aggfunc='count')
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))  # Create a stacked bar chart
plt.title('Python - Stacked Bar Chart - Count of Passengers by Class and Sex')  # Set the title of the chart
plt.xlabel('Class')  # Set the label for the x-axis
plt.ylabel('Count of Passengers')  # Set the label for the y-axis
plt.savefig('Python_Stacked_Bar_Chart.png')  # Save the figure as a PNG file
plt.show()  # Display the chart
print("Stacked bar chart created and saved as 'Python_Stacked_Bar_Chart.png'!")
```

```python
# Step 5: Create a Pie Chart
print("Step 5: Creating a pie chart...")
plt.figure(figsize=(8, 8))  # Set the size of the figure
# Create a pie chart for the 'Pclass' column
data['Pclass'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Python - Pie Chart - Distribution of Passengers by Class')  # Set the title of the chart
plt.ylabel('')  # Remove the y-axis label
plt.savefig('Python_Pie_Chart.png')  # Save the figure as a PNG file
plt.show()  # Display the chart
print("Pie chart created and saved as 'Python_Pie_Chart.png'!")
```

```python
# Step 6: Create a Donut Chart
print("Step 6: Creating a donut chart...")
plt.figure(figsize=(8, 8))  # Set the size of the figure
# Create a pie chart for the 'Pclass' column with a hole in the middle to make it a donut chart
sizes = data['Pclass'].value_counts()
colors = ['skyblue', 'lightgreen', 'lightcoral']
plt.pie(sizes, labels=sizes.index, colors=colors, startangle=90, autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.title('Python - Donut Chart - Distribution of Passengers by Class')  # Set the title of the chart
plt.savefig('Python_Donut_Chart.png')  # Save the figure as a PNG file
plt.show()  # Display the chart
print("Donut chart created and saved as 'Python_Donut_Chart.png'!")
```

```python
# Step 7: Create a Line Chart
print("Step 7: Creating a line chart...")
plt.figure(figsize=(10, 6))  # Set the size of the figure
sns.lineplot(data=data, x='Age', y='Fare', hue='Pclass')  # Create a line plot for 'Age' vs 'Fare' grouped by 'Pclass'
plt.title('Python - Line Chart - Fare vs Age by Class')  # Set the title of the chart
plt.xlabel('Age')  # Set the label for the x-axis
plt.ylabel('Fare')  # Set the label for the y-axis
plt.savefig('Python_Line_Chart.png')  # Save the figure as a PNG file
plt.show()  # Display the chart
print("Line chart created and saved as 'Python_Line_Chart.png'!")
```
