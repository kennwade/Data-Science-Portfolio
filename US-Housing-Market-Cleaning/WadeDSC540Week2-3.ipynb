**Matplotlib Line Chart Creation:**

**Steps:**

   I utilized Matplotlib, a powerful data visualization library, to create a line chart. The real-world population data provided in the `world-population.xlsm` file served as the basis for this visualization.

   - First, I loaded the data from the Excel file and stored it in a DataFrame.

   - I customized the chart by adding a title, labels for the x and y axes, gridlines, and adjusted the layout for better readability.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the Excel file
file_path = "/Users/kennwade/Downloads/world-population.xlsm"
data = pd.read_excel(file_path)

# Extract relevant columns for plotting
years = data['Year']  # Extracting the 'Year' column
population = data['Population']  # Extracting the 'Population' column

# Plotting
plt.figure(figsize=(10, 6))  # Setting the figure size
plt.plot(years, population, marker='o', color='b', linestyle='-')  # Plotting the data
plt.title('World Population Growth')  # Adding title to the plot
plt.xlabel('Year')  # Adding label for x-axis
plt.ylabel('Population')  # Adding label for y-axis
plt.grid(True)  # Adding gridlines to the plot
plt.xticks(rotation=45)  # Rotating x-axis labels for better readability
plt.tight_layout()  # Adjusting layout to prevent clipping of labels
plt.show()  # Displaying the plot
```

**Data Wrangling Activities: Activity 1.01 (Page 23) Handling Lists**

**Steps:**

1. Create a list of 100 random numbers.
2. Create a new list from this random list, containing only numbers that are divisible by 3.
3. Calculate the length of these two lists and store the difference in a new variable.
4. Use a loop to repeat steps 1, 2, and 3, and find the difference variable 10 times.
5. Calculate the arithmetic mean of these 10 difference values

```python
import random

# Initialize a list to store the differences
difference_values = []

# Repeat the experiment 10 times
for _ in range(10):
    # Step 1: Create a list of 100 random numbers
    random_list = [random.randint(1, 1000) for _ in range(100)]
    
    # Step 2: Create a new list with numbers divisible by 3
    divisible_by_3_list = [num for num in random_list if num % 3 == 0]
    
    # Step 3: Calculate the length of the two lists and store the difference
    length_random_list = len(random_list)
    length_divisible_by_3_list = len(divisible_by_3_list)
    difference = length_random_list - length_divisible_by_3_list
    
    # Step 4: Store the difference in the list
    difference_values.append(difference)

# Step 5: Find the arithmetic mean of the difference values
average_difference = sum(difference_values) / len(difference_values)

# Print the output
print("Average Difference between the lengths of the two lists:", average_difference)
```

**Data Wrangling Activities: Activity 1.02 (Page 43) Analyzing a Multiline String and Generating the Unique Word Count**

**Steps:**

1. I created a multiline_text variable by copying the text from the first chapter of Pride and Prejudice from the downloaded GitHub data.  
2. I found the type and length of the multiline_text string using the type and len commands.
3. I removed all new lines and symbols using the replace method.
4. I found all of the words in multiline_text using the split method.
5. I created a list from this list that contained only the unique words.
6. I counted the number of times each unique word appeared in the list using the key and value in dict.
7. I found the top 25 words from the unique words that I had found using the slice method.

```python
# Step 1: Create a multiline_text variable by reading the text from the file
file_path = "/Users/kennwade/Downloads/P&P Ch01.txt"
with open(file_path, "r") as file:
    multiline_text = file.read()

# Step 2: Find the type and length of the multiline_text string
multiline_text_type = type(multiline_text)
multiline_text_length = len(multiline_text)

# Step 3: Remove all new lines and symbols using the replace method
cleaned_text = multiline_text.replace("\n", "").replace(",", "").replace(".", "").replace(":", "").replace(";", "").replace("!", "").replace("?", "")

# Step 4: Find all words in the cleaned_text using the split method
words = cleaned_text.split()

# Step 5: Create a list of unique words
unique_words = list(set(words))

# Step 6: Count the occurrences of each unique word
word_count = {}
for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

# Step 7: Find the top 25 words from the unique words
top_25_words = sorted(word_count.items(), key=lambda x: x[1], reverse=True)[:25]

# Print the output
print("Type of multiline_text:", multiline_text_type)
print("Length of multiline_text:", multiline_text_length)
print("Top 25 words:")
for word, count in top_25_words:
    print(f"{word}: {count}")
```

**Data Wrangling Activities: Activity 2.01 (Page 70) Permutation, Iterator, Lambda, and List**

**Steps:**
1. I looked up the definitions of permutations and dropwhile from itertools.
2. I wrote an expression to generate all possible three-digit numbers using 0, 1, and 2.
3. I looped over the iterator expression I generated before, printed each element returned by the iterator, and used assert and isinstance to ensure the elements are of tuple type.
4. I wrote the loop again, using dropwhile with a lambda expression to drop any leading zeros from the tuples. I casted the output of dropwhile to a list.
5. I checked the actual type that dropwhile returns.
6. I combined the preceding code into one block, this time writing a separate function where I passed the list generated from dropwhile, and the function returned the whole number contained in the list. I treated the incoming list as a stack in the function and generated the number by reading the individual digits from the stack.


```python
from itertools import permutations, dropwhile

# Step 1: Generate all possible three-digit numbers using 0, 1, and 2
numbers_permutations = permutations([0, 1, 2], 3)

# Step 2 & 3: Loop over the iterator, print each element, and check their type
for number_tuple in numbers_permutations:
    print(number_tuple)
    assert isinstance(number_tuple, tuple)

# Step 4: Use dropwhile and lambda to drop leading zeros from tuples
cleaned_numbers = [list(dropwhile(lambda x: x == 0, number_tuple)) for number_tuple in numbers_permutations]

# Step 5: Check the type returned by dropwhile
print(type(cleaned_numbers))

# Step 6: Write a function to convert the list to a whole number
def list_to_number(lst):
    number = 0
    for digit in lst:
        number = number * 10 + digit
    return number

# Test the function with cleaned_numbers
for number_list in cleaned_numbers:
    print(list_to_number(number_list))
```

**Data Wrangling Activities: Activity 2.02 (Page 81) Designing My Own CSV Parser**

**Steps:**

1. **Importing Necessary Modules and Setting Up:**
   - I imported `zip_longest` from `itertools`.
   - I created a function to zip header, line, and fillvalue=None.
   - I opened the accompanying `sales_record.csv` file from the provided GitHub link using 'r' mode inside a `with` block and ensured that it was opened.

2. **Extracting Column Names:**
   - I read the first line of the CSV file.
   - I used string methods to generate a list of all the column names.

3. **Reading CSV File Line by Line:**
   - I started reading the file line by line.

4. **Processing Each Line:**
   - For each line, I passed it to a function along with the list of headers.
   - The function constructed a dictionary from the line and headers, filling in missing values with `None`.

```python
from itertools import zip_longest

# Step 1: Define a function to create a dictionary from CSV line
def return_dict_from_csv_line(header, line):
    # Zip header and line, then construct dictionary
    zipped_line = zip_longest(header, line, fillvalue=None)
    ret_dict = {kv[0]: kv[1] for kv in zipped_line}
    return ret_dict  

# Step 2: Open the CSV file and read the first line to get column names
with open("/Users/kennwade/Downloads/sales_record.csv", "r") as fd:
    first_line = fd.readline()
    header = first_line.replace("\n", "").split(",")

    # Step 3: Read the file line by line
    for i, line in enumerate(fd):
        # Step 4: Process each line to create a dictionary
        line = line.replace("\n", "").split(",")
        data_dict = return_dict_from_csv_line(header, line)
        # Print the dictionary representing each line of data
        print(f"Processed data from line {i + 2}: {data_dict}")
        # Break after printing the first 10 lines
        if i > 10:
            break
```
